Prophet follows the <span style="color:green">sklearn</span> model API. We create an instance of the <span style="color:green">Prophet</span> class and then call its <span style="color:green">fit</span> and <span style="color:green">predict</span> methods.

The input to Prophet is always a dataframe with two columns: <span style="color:green">ds</span> and <span style="color:green">y</span>. The <span style="color:green">ds</span> (datestamp) column must contain a date or datetime (either is fine). The <span style="color:green">y</span> column must be numeric, and represents the measurement we wish to forecast.

As an example, let’s look at a time series of daily page views for the Wikipedia page for [Peyton Manning](https://en.wikipedia.org/wiki/Peyton_Manning). We scraped this data using the Wikipediatrend package in R. Peyton Manning provides a nice example because it illustrates some of Prophet’s features, like multiple seasonality, changing growth rates, and the ability to model special days (such as Manning’s playoff and superbowl appearances). The CSV is available here.

First we’ll import the data and log-transform the <span style="color:green">y</span> variable.

```python
import pandas as pd
import numpy as np
from fbprophet import Prophet

# Disable warning from Numpy float conversion
import warnings
warnings.filterwarnings(action='once')
# ignore all warnings
# warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('example_wp_peyton_manning.csv')
df['y'] = np.log(df['y'])
df.head()
```

We fit the model by instantiating a new <span style="color:green">Prophet</span> object. Any settings to the forecasting procedure are passed into the constructor. Then you call its <span style="color:green">fit</span> method and pass in the historical dataframe. Fitting should take 1-5 seconds.

```python
m = Prophet(daily_seasonality=True) # By default, daily_seasonality=False
m.fit(df)
```

Predictions are then made on a dataframe with a column <span style="color:green">ds</span> containing the dates for which a prediction is to be made. You can get a suitable dataframe that extends into the future a specified number of days using the helper method <span style="color:green">Prophet.make_future_dataframe</span>. By default it will also include the dates from the history, so we will see the model fit as well.

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```

The <span style="color:green">predict</span> method will assign each row in <span style="color:green">future</span> a predicted value which it names yhat. If you pass in historical dates, it will provide an in-sample fit. The <span style="color:green">forecast</span> object here is a new dataframe that includes a column <span style="color:green">yhat</span> with the forecast, as well as columns for components and uncertainty intervals.

```python
forecast = m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

You can plot the forecast by calling the <span style="color:green">Prophet.plot</span> method and passing in your forecast dataframe.

```python
m.plot(forecast)
```

If you want to see the forecast components, you can use the <span style="color:green">Prophet.plot_components</span> method. By default you’ll see the trend, yearly seasonality, and weekly seasonality of the time series. If you include holidays, you’ll see those here, too.

```python
m.plot_components(forecast)
```
